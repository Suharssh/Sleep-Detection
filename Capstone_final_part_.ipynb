Feature Extraction and model training

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Masking
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2


def downsample_signals_chunked(df_path, time_col="Time", window_size=1000, chunk_size=50000):
    downsampled_chunks = []
    for chunk in pd.read_csv(df_path, chunksize=chunk_size, header=None):
        chunk = chunk.drop(columns=[1])
        chunk.rename(columns={0: "Time"}, inplace=True)
        chunk["Time"] = chunk["Time"] // window_size * window_size
        downsampled_chunk = chunk.groupby("Time").mean().reset_index()
        downsampled_chunks.append(downsampled_chunk)
    return pd.concat(downsampled_chunks, ignore_index=True)

df2_downsampled = downsample_signals_chunked(file_path2, window_size=1000)

def extract_sequences(df1, df2_downsampled):
    X_seq, y_seq = [], []

    for _, row in df1.iterrows():
        start, duration = row["Start_Time"], row["Duration"]
        end = start + duration

        mask = (df2_downsampled["Time"] >= start) & (df2_downsampled["Time"] <= end)
        seq_X = df2_downsampled[mask].drop(columns=["Time"]).values

        if len(seq_X) > 0:
            X_seq.append(seq_X)
            y_seq.append(row["Sleep_Stage"])

    return np.array(X_seq, dtype=object), np.array(y_seq)

X_seq, y_seq = extract_sequences(df1, df2_downsampled)

label_encoder = LabelEncoder()
y_seq_encoded = label_encoder.fit_transform(y_seq)
y_seq_onehot = to_categorical(y_seq_encoded)

X_seq_padded = pad_sequences(X_seq, dtype='float32', padding='post', value=0.0)

X_train, X_test, y_train, y_test = train_test_split(X_seq_padded, y_seq_onehot, test_size=0.2, random_state=42)

input_shape = (X_train.shape[1], X_train.shape[2])
num_classes = y_train.shape[1]

model = Sequential([
    Masking(mask_value=0.0, input_shape=input_shape),
    LSTM(32, return_sequences=True, kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.3),
    LSTM(16, kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(16, activation='relu', kernel_regularizer=l2(0.001)),
    Dense(num_classes, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.0005), metrics=['accuracy'])

lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3)

history = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                    epochs=100, batch_size=32,
                    callbacks=[lr_scheduler])

test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
y_pred = model.predict(X_test)

y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

unique_classes = np.unique(np.concatenate((y_true_classes, y_pred_classes)))

target_names = [label_encoder.classes_[i] for i in unique_classes]

report = classification_report(y_true_classes, y_pred_classes, target_names=target_names, labels=unique_classes)
print(report)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)
cm = confusion_matrix(y_true_classes, y_pred_classes)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
